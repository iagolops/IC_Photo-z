````python
from dblinea import DBBase
from astroquery.vizier import Vizier
import matplotlib.pyplot as plt
from astropy import units as u
from astropy.coordinates import SkyCoord
from tqdm import tqdm
from PIL import Image
import numpy as np
import astropy 
import pandas as pd
````

## Selecionando as colunas do catálogo VIPERS-W1

````python
catalog_name = 'J/A+A/609/A84/vipersw1' #VIPERS catalog
vizier = Vizier(columns = ['ID','RAJ2000','DEJ2000','zsp','zflg','SSR','classFlag'],row_limit=-1)
catalog = vizier.get_catalogs(catalog_name)
````

````python
table = catalog[catalog_name]
table
````

### Usando as flags de boa confiança como máscaras para filtrar o catálogo

````python
mask =( (table['zflg']>3) & (table['zflg']<5))
vipers_df=table[mask]
vipers_df
````

````python
db=DBBase()
schema = "des_dr2"  
tablename = "coadd_objects"
````

````python
db.get_table_columns(tablename, schema=schema)
````

## Selecionando os dados do DES_DR2 que estão, na mesma área do céu, do VIPERS-W1

```sql
SELECT tilename,extended_class_coadd, coadd_object_id ,alphawin_j2000 ,deltawin_j2000 ,mag_auto_g ,mag_auto_r ,mag_auto_i ,mag_auto_z ,mag_auto_y, magerr_auto_g, magerr_auto_r, magerr_auto_i, magerr_auto_z, magerr_auto_y
FROM des_dr2.coadd_objects 
WHERE q3c_poly_query(ra, dec, ARRAY[{xlim[30]}, {ylim[-6]}, {xlim[30]}, {ylim[-4]}, {xlim[39]}, {ylim[-4]}, {xlim[39]}, {ylim[-6]}])
AND extended_class_coadd>=2
limit 2000000
```

````python


query = 'SELECT tilename, extended_class_coadd, coadd_object_id ,alphawin_j2000 ,deltawin_j2000 ,mag_auto_g ,mag_auto_r ,mag_auto_i ,mag_auto_z ,mag_auto_y, magerr_auto_g, magerr_auto_r, magerr_auto_i, magerr_auto_z, magerr_auto_y FROM des_dr2.coadd_objects WHERE q3c_poly_query(ra, dec, ARRAY[30, -6, 30, -4, 39, -4, 39, -6]) AND extended_class_coadd>=2 limit 2000000'
````

````python
%%time
des_df = db.fetchall_df(query) #DES_DR2 catalog
des_df = des_df[des_df['mag_auto_i']<=24]
des_df 
````

### Plot da região no céu com os objetos selecionados

````python
plt.scatter(des_df['alphawin_j2000'], des_df['deltawin_j2000'], color = 'blue', label = 'DES',s=0.2)
plt.scatter(vipers_df['RAJ2000'],vipers_df['DEJ2000'],color = 'red', label = 'VIPERS', s=0.2)
plt.legend(loc='best')
````

## Removendo os dados do VIPERS que estão com uma distância maior do que 1 arcsec dos dados do DR2

````python
%%time
values = [] #bin list
for object_VIPERS in vipers_df:
    #mask is a square 
    near_objects = des_df[(des_df['alphawin_j2000']<(object_VIPERS['RAJ2000']+0.0003)) & (des_df['alphawin_j2000']>(object_VIPERS['RAJ2000']-0.0003)) & (des_df['deltawin_j2000']<(object_VIPERS['DEJ2000']+0.0003)) & (des_df['deltawin_j2000']>(object_VIPERS['DEJ2000']-0.0003))]
    if len(near_objects) <1:
        values.append(0)
    elif len(near_objects) >=1:
        values.append(1)
````

````python
vipers_df['values'] = values #bin list where 1 equals to retained and 0 equals to removed
retained = vipers_df[vipers_df['values']==1]
removed = vipers_df[vipers_df['values']==0]

print(len(removed)) #number of removed objects
print(len(retained))
print(len(vipers_df))
````

````python
vipers_df = vipers_df[vipers_df['values']==1]
vipers_df
````

## (Opcional) Teste verificando os valores que estão dentro do quadrado, porém com distancia maior que 1 arcsec (fora do círculo)

````python
%%time
values_2 = []
for object_VIPERS in vipers_df:
    near_objects = des_df[(des_df['alphawin_j2000']<(object_VIPERS['RAJ2000']+0.0003)) & (des_df['alphawin_j2000']>(object_VIPERS['RAJ2000']-0.0003)) & (des_df['deltawin_j2000']<(object_VIPERS['DEJ2000']+0.0003)) & (des_df['deltawin_j2000']>(object_VIPERS['DEJ2000']-0.0003))]
    point = SkyCoord(ra = object_VIPERS['RAJ2000'], dec = object_VIPERS['DEJ2000'] , unit=[u.deg,u.deg])
    points = SkyCoord(ra = near_objects['alphawin_j2000'], dec = near_objects['deltawin_j2000'], unit=[u.deg,u.deg])
    separation = point.separation(points)
    for element in separation:
        dist = astropy.coordinates.Angle(element)
        deg = dist.to('degree')
        if deg <= astropy.coordinates.Angle(1, unit='arcsec'):
            values_2.append(1)
        else:
            values_2.append(0)
 
print(0 in values_2)
````

## Descobrindo a equivalência de objetos entre os dois catálogos e adicionando o redshift espectroscópico (zs) do VIPERS no data frame do DES-DR2

````python
%%time
des_df['zs'] = 0.0
for object_VIPERS in vipers_df:
    near_objects_indice = (des_df['alphawin_j2000']<(object_VIPERS['RAJ2000']+0.0001)) & (des_df['alphawin_j2000']>(object_VIPERS['RAJ2000']-0.0001)) & (des_df['deltawin_j2000']<(object_VIPERS['DEJ2000']+0.0001)) & (des_df['deltawin_j2000']>(object_VIPERS['DEJ2000']-0.0001))            
    near_objects = des_df[near_objects_indice]
    
    #these lines was for bigger intevals wich can have more than 1 object in the square
    
    #point = SkyCoord(ra = object_VIPERS['RAJ2000'], dec = object_VIPERS['DEJ2000'] , unit=[u.deg,u.deg])
    #points = SkyCoord(ra = near_objects['alphawin_j2000'], dec = near_objects['deltawin_j2000'], unit=[u.deg,u.deg])
    #separation = points.separation(point)
    
    #need to take the min separation and use his index 
    
    des_df.loc[near_objects_indice, 'zs'] = object_VIPERS['zsp']
    
````

## Organizando o data frame de maneira que fique parecido com os dados de treino do TPZ

````python
data_sample_temp = des_df[des_df['zs']>0] #selecting only the objects wich have z-spec
````

````python
data_sample = data_sample_temp.drop(columns=['extended_class_coadd','tilename','coadd_object_id','deltawin_j2000','alphawin_j2000'])
````

````python
data_sample.rename(columns={'mag_auto_g':'g', 'mag_auto_r':'r', 'mag_auto_i':'i', 'mag_auto_z':'z', 'mag_auto_y':'y'}, inplace = True)
data_sample.rename(columns={'magerr_auto_g':'eg', 'magerr_auto_r':'er', 'magerr_auto_i':'ei', 'magerr_auto_z':'ez', 'magerr_auto_y':'ey'}, inplace = True)
````

````python
zs = data_sample.pop('zs') #putting the z-spec first
data_sample.insert(0, 'zs', zs)
````

````python
data_sample['g-r'] = data_sample['g'] - data_sample['r']
data_sample['r-i'] = data_sample['r'] - data_sample['i']
data_sample['i-z'] = data_sample['i'] - data_sample['z']
data_sample['z-y'] = data_sample['z'] - data_sample['y']
data_sample['eg-r'] = data_sample['eg'] - data_sample['er']
data_sample['er-i'] = data_sample['er'] - data_sample['ei']
data_sample['ei-z'] = data_sample['ei'] - data_sample['ez']
data_sample['ez-y'] = data_sample['ez'] - data_sample['ey']
````

````python
data_sample
````

## Separando a amostra entre arquivos de treino e de teste e salvando

````python
data_sample_train = data_sample.sample(frac=0.5, random_state=42)
data_sample_test = data_sample.drop(data_sample_train.index)
````
