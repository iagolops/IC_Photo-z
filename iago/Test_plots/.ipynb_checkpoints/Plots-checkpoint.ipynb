# Mean bias

```python
bins = np.linspace(0.02, np.max(rzphot), 9)
mean_r=[]

for index in range(len(bins) - 1):
    bin_lower = bins[index]
    bin_upper = bins[index + 1]
    
    values_r = rzphot[(rzphot >= bin_lower) & (rzphot <= bin_upper)]
    values_s = rzspec[(rzphot >= bin_lower) & (rzphot <= bin_upper)]
    
    
    mean_r.append(np.mean(values_r - values_s))
     

        
plt.axhline(0,color='k',linestyle='--',alpha=0.5)
plt.plot(bins[:-1],mean_r,'bo-',label='Reg')
plt.xlabel('Zphot')
plt.ylabel('Δz ')
plt.legend(loc='upper right',framealpha=1.0)
```

# Plot sigma 68

```python
bins = np.linspace(0.02, np.max(rzphot), 9)
mean_r=[]

for index in range(len(bins) - 1):
    bin_lower = bins[index]
    bin_upper = bins[index + 1]
    
    values_r = rzphot[(rzphot >= bin_lower) & (rzphot <= bin_upper)]
    values_s = rzspec[(rzphot >= bin_lower) & (rzphot <= bin_upper)]


    deltabias = (values_r-values_s)
    mean_bias = np.sum((values_r-values_s)/len(values_r))

    
    sigma = np.std(deltabias) #sigma68 for each bin
    
    mean_r.append(sigma) 
     

        

plt.plot(bins[:-1],mean_r,'ko-')
plt.xlabel('Zphot')
plt.ylabel(r'$σ_{68}$')
#plt.legend(loc='upper right',framealpha=1.0)
```

# Outliers

```python
bins = np.linspace(0.02, np.max(rzphot), 9)
fraction_outliers = []
sigmaz = []
meanz = []
outliers_2 = []

for index in range(len(bins) - 1):
    bin_lower = bins[index]
    bin_upper = bins[index + 1]
    
    values_r = rzphot[(rzphot >= bin_lower) & (rzphot <= bin_upper)]
    values_s = rzspec[(rzphot >= bin_lower) & (rzphot <= bin_upper)]

    deltabias = (values_r - values_s)
    mean_bias = np.mean(deltabias)  # Mean bias for each bin
    meanz.append(mean_bias)
    
    sigma = np.std(deltabias)  # Standard deviation (sigma) for each bin
    sigmaz.append(sigma)

    # Calculate the fraction of outliers outside 3 sigma
    outliers = np.abs(deltabias) > 3 * sigma
    fraction_outlier = np.sum(outliers) / len(outliers)
    
    fraction_outliers.append(fraction_outlier)
    
    #2 sigma
    outliers2 = np.abs(deltabias) > 2 * sigma
    fraction_outlier2 = np.sum(outliers2) / len(outliers2)
    
    
    outliers_2.append(fraction_outlier2)



fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True)
plt.subplots_adjust(hspace=0.4)  # Ajuste o espaço entre os subplots

# Configure os limites x comuns para todos os subplots
x_lim = (0, np.max(rzphot))

# Subplot 1: Mean Bias
axes[0, 0].plot(bins[:-1], meanz, 'bo-')
axes[0, 0].axhline(0, color='black', linestyle='--')
axes[0, 0].set_ylabel(r'$\Delta z$', fontsize=20)
axes[0, 0].set_xlim(x_lim)
axes[0, 0].tick_params(axis='both', labelsize=14)
axes[0, 0].grid(True)

# Subplot 2: Standard Deviation
axes[0, 1].plot(bins[:-1], sigmaz, 'go-')
axes[0, 1].set_ylabel(r'$\sigma_{68}$', fontsize=20)
axes[0, 1].set_xlim(x_lim)
axes[0, 1].set_ylim(0,max(sigmaz)+0.01)
axes[0, 1].tick_params(axis='both', labelsize=14)
axes[0, 1].grid(True)

# Subplot 3: Outliers
axes[1, 0].plot(bins[:-1], fraction_outliers, 'ro-')
axes[1, 0].set_xlabel(r'$Z_{phot}$', fontsize=20)
axes[1, 0].set_ylabel('out$_{3σ}$', fontsize=20)
axes[1, 0].set_xlim(x_lim)
axes[1, 0].set_ylim(0,max(fraction_outliers)+0.01)
axes[1, 0].tick_params(axis='both', labelsize=14)
axes[1, 0].grid(True)

# Subplot 4: 
axes[1, 1].plot(bins[:-1],outliers_2,'o-',color='darkorange')
axes[1, 1].set_xlabel(r'$Z_{phot}$', fontsize=20)
axes[1, 1].set_ylabel('out$_{2σ}$', fontsize=20)
axes[1, 1].set_xlim(x_lim)
axes[1, 1].set_ylim(0,max(outliers_2)+0.01)
axes[1, 1].tick_params(axis='both', labelsize=14)
axes[1, 1].grid(True)
axes[1, 1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

# Ajuste o layout dos subplots
plt.tight_layout()

# Salve a figura em um arquivo (opcional)
if classification:
    plt.savefig(f'/home/iago/ic_cosmo/plots/class_metrics_{survey}.png')
else:
    plt.savefig(f'/home/iago/ic_cosmo/plots/metrics_{survey}.png')

# Exiba a figura
plt.show()

```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True)
plt.subplots_adjust(hspace=0.4)  # Ajuste o espaço entre os subplots

for test in test_list:
    column_names = ['ztrue', 'zmode0', 'zmean1', 'zConf0', 'zConf1', 'err0', 'err1']
    if classification:
        df_temp = pd.read_csv(f'/home/iago/Plots-TPZ/iago/ml_analyses/results/results.{test}.1.mlz', names=column_names, header=None, skiprows=1, sep=' ')
    else:   
        df_temp = pd.read_csv(f'/home/iago/Plots-TPZ/iago/ml_analyses/results/results.{test}.0.mlz', names=column_names, header=None, skiprows=1, sep=' ')

    zphot = df_temp['zmean1']
    zspec = df_temp['ztrue']

    bins = np.linspace(0.02, np.max(zphot), 9)
    fraction_outliers = []
    sigmaz = []
    meanz = []
    outliers_2 = []

    for index in range(len(bins) - 1):
        bin_lower = bins[index]
        bin_upper = bins[index + 1]

        values_r = zphot[(zphot >= bin_lower) & (zphot <= bin_upper)]
        values_s = zspec[(zphot >= bin_lower) & (zphot <= bin_upper)]

        deltabias = (values_r - values_s)
        mean_bias = np.mean(deltabias)  # Mean bias for each bin
        meanz.append(mean_bias)

        sigma = np.std(deltabias)  # Standard deviation (sigma) for each bin
        sigmaz.append(sigma)

        # Calculate the fraction of outliers outside 3 sigma
        outliers = np.abs(deltabias) > 3 * sigma
        fraction_outlier = np.sum(outliers) / len(outliers)

        fraction_outliers.append(fraction_outlier)

        # 2 sigma
        outliers2 = np.abs(deltabias) > 2 * sigma
        fraction_outlier2 = np.sum(outliers2) / len(outliers2)

        outliers_2.append(fraction_outlier2)

    # Configure os limites x comuns para todos os subplots
    x_lim = (0, np.max(zphot))

    # Subplot 1: Mean Bias
    axes[0, 0].plot(bins[:-1], meanz, 'o-', label=test)

    # Subplot 2: Standard Deviation
    axes[0, 1].plot(bins[:-1], sigmaz, 'o-', label=test)

    # Subplot 3: Outliers
    axes[1, 0].plot(bins[:-1], fraction_outliers, 'o-', label=test)

    # Subplot 4: Outliers_2
    axes[1, 1].plot(bins[:-1], outliers_2, 'o-', label=test)

# Configurar rótulos e legendas
axes[0, 0].axhline(0, color='black', linestyle='--')
axes[0, 0].set_ylabel(r'$\Delta z$', fontsize=20)
axes[0, 0].set_xlim(x_lim)
axes[0, 0].set_ylim(-0.2,0.2)
axes[0, 0].tick_params(axis='both', labelsize=14)
axes[0, 0].grid(True)

axes[0, 1].set_ylabel(r'$\sigma_{68}$', fontsize=20)
axes[0, 1].set_xlim(x_lim)
axes[0, 1].tick_params(axis='both', labelsize=14)
axes[0, 1].grid(True)

axes[1, 0].set_xlabel(r'$Z_{phot}$', fontsize=20)
axes[1, 0].set_ylabel('out$_{3σ}$', fontsize=20)
axes[1, 0].set_xlim(x_lim)
axes[1, 0].set_ylim(0,0.1)
axes[1, 0].tick_params(axis='both', labelsize=14)
axes[1, 0].grid(True)

axes[1, 1].set_xlabel(r'$Z_{phot}$', fontsize=20)
axes[1, 1].set_ylabel('out$_{2σ}$', fontsize=20)
axes[1, 1].set_xlim(x_lim)
axes[1, 1].set_ylim(0,0.25)
axes[1, 1].tick_params(axis='both', labelsize=14)
axes[1, 1].grid(True)
axes[1, 1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

# Adicione uma legenda para cada linha
for ax in axes.ravel():
    ax.legend()

# Ajuste o layout dos subplots
plt.tight_layout()

# Salve a figura em um arquivo (opcional)

if classification:
    plt.savefig(f'/home/iago/ic_cosmo/plots/class_metrics_combined.png')
else:
    plt.savefig(f'/home/iago/ic_cosmo/plots/metrics_combined.png')

# Exiba a figura
plt.show()
```
